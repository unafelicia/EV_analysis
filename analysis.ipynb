```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# superchager capacity vs ev sales (all brand)

x = df["QuarterLabel"]
y1 = df["ALL_BRAN"]
y2 = df["supercharger"]

# 标准化
y1 = y1 / y1.max()
y2 = y2 / y2.max()

plt.figure(figsize=(10, 6))
plt.plot(x, y1, marker='o', label='ALL_BRAN')
plt.plot(x, y2, marker='s', label='supercharger')
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.title("EV Sales (All Brands) vs Supercharger Count")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()
plt.show()
```

```python
# ev sales (audi only)
x = df["QuarterLabel"]
y1 = df["Audi"]
y2 = df["supercharger"]

plt.figure(figsize=(10, 6))
plt.plot(x, y1, marker='o', label='Audi')
# plt.plot(x, y2, marker='s', label='SuperCharger_N')
plt.xlabel("Quarter")
plt.ylabel("Sales")
plt.title("EV Sales (Audi Only)")
plt.xticks(rotation=45)
plt.legend()
plt.grid()
plt.tight_layout()

for label, quarter in audi_events.items():
    if quarter in x.values:
        xpos = list(x).index(quarter)
        plt.axvline(x=quarter, color="red", linestyle="--", alpha=0.7)
        mid_y = (plt.ylim()[0] + plt.ylim()[1]) / 2
        plt.text(xpos, mid_y, label, rotation=90, color="red",
                 ha="center", va="center", fontsize=9, bbox=dict(facecolor="white", alpha=0.6, edgecolor="none"))
plt.show()
```
